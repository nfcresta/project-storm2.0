# Initial Data Cleansing

```python
#Import Dependencies
import pandas as pd
import datetime as dt
```

```python
#Import the CSVs

storms_18 = pd.read_csv("data files/Storms_2018.csv")
storms_19 = pd.read_csv("data files/Storms_2019.csv")
storms_20 = pd.read_csv("data files/Storms_2020.csv")

storms_18.head()
```

```python
#Merging the Data (Merging 2018 and 2019)

storm_merge1 = pd.merge(storms_18, storms_19, how = "outer")
storm_merge1.head()
```

```python
#Merging the Data (Merging with 2020)
storm_merge = pd.merge(storm_merge1, storms_20, how = "outer")

storm_merge.head()
```

```python
storm_merge.columns
```

```python
#Dropping Unneccessary Columns
storms_merge = storm_merge.drop(columns = ['state_fips_code','cz_type', 'cz_fips_code', 'cz_name', 'source', 'magnitude',
       'magnitude_type', 'flood_cause', 'tor_f_scale', 'tor_length',
       'tor_width', 'tor_other_wfo', 'location_index', 'event_range',
       'event_azimuth', 'reference_location', 'event_latitude',
       'event_longitude', 'event_point'])

storms_merge.head(20)
```

```python
#Converting to a legible datetime format

storms_merge['event_begin_time'] = pd.to_datetime(storms_merge['event_begin_time'])
storms_merge['event_end_time'] = pd.to_datetime(storms_merge['event_end_time'])
```

```python
#Extracting the Month
storms_merge['event_month'] = storms_merge['event_begin_time'].dt.month
```

```python
#Extracting the Year
storms_merge['event_year'] = storms_merge['event_begin_time'].dt.year

storms_merge.head()
```

```python
#Dropping NAs
storms_merge = storms_merge.dropna()
storms_merge.shape
```

```python
#Dropping Obsolete Columns
storms_merge = storms_merge.drop(columns = ['event_begin_time','event_end_time'])

storms_merge.head()
```

# Pre-Processing

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(storms_merge['state'])
encoded_state = label_encoder.transform(storms_merge['state'])

label_encoder = LabelEncoder()
label_encoder.fit(storms_merge['event_type'])
encoded_event_type = label_encoder.transform(storms_merge['event_type'])

label_encoder = LabelEncoder()
label_encoder.fit(storms_merge['wfo'])
encoded_wfo = label_encoder.transform(storms_merge['wfo'])

label_encoder = LabelEncoder()
label_encoder.fit(storms_merge['event_timezone'])
encoded_event_timezone = label_encoder.transform(storms_merge['event_timezone'])

encoded_event_timezone.shape
```

```python
# Create variables
dropped_df = storms_merge.drop(columns=['episode_id', 'event_id', 'damage_property', 'state', 'event_type', 'wfo', 'event_timezone'])

dropped_df['state'] = encoded_state
dropped_df['event_type'] = encoded_event_type
dropped_df['wfo'] = encoded_wfo
dropped_df['event_timezone'] = encoded_event_timezone
```

```python
# Set X and Y variable (Value Reshaping)
X = dropped_df
y = storms_merge['damage_property'].values.reshape(-1, 1)
X
```

# Training and Fitting

```python
# Establishing train-test and split data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
#Confirming we're set to move forward
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data

X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)
```

```python
#Transforming scaled data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
#Executing Linear Regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train_scaled)
```

```python
# Printing Training and Testing Model Scores
print(f"Training Data Score: {model.score(X_train_scaled, y_train_scaled)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test_scaled)}")

# Use our model to predict a value
predicted = model.predict(X_test_scaled)
```

```python
#Working with RandomForestClassifier and tree
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
